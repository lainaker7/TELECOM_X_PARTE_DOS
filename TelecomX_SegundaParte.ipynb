## üîç **Analizar la utilidad de la informaci√≥n**


Para continuar con el an√°lisis y asegurarnos de que todas las columnas son √∫tiles, vamos a revisar los tipos de datos y la cantidad de valores √∫nicos de cada columna. Esto nos ayudar√° a identificar posibles candidatos para una limpieza o transformaci√≥n posterior.

```python
print('Informaci√≥n general del DataFrame:')
df.info()

print('\nN√∫mero de valores √∫nicos por columna:')
print(df.nunique())
```

```python
# Eliminamos la columna ID de cliente ya que tiene valores unicos  y no aporta valor a los modelos predictivos

df = df.drop('ID_cliente', axis=1)
display(df.head())
```

```python
print('Informaci√≥n general del DataFrame despu√©s de las eliminaciones:')
df.info()

print('\nN√∫mero de valores √∫nicos por columna despu√©s de las eliminaciones:')
print(df.nunique())
```

 ## **Transformamos Variables con ONE-HOT-ENCODING**‚úèÔ∏è

```python
df_encoded = pd.get_dummies(df, drop_first=False)
display(df_encoded.head())
```

```python
print('Informaci√≥n general del DataFrame despu√©s de la codificaci√≥n:')
df_encoded.info()
```

Se calcula las proporciones de clientes con y sin deserci√≥n. Si el desbalance es significativo, puede ser necesario aplicar t√©cnicas de manejo de desbalance de clases (como sobremuestreo, submuestreo o ajuste de pesos de clase) al entrenar los modelos predictivos para evitar que el modelo se incline demasiado hacia la clase mayoritaria

```python
churn_counts = df_encoded['Deserci√≥n_S√≠'].value_counts()
no_churn_counts = df_encoded['Deserci√≥n_No'].value_counts()

total_customers = len(df_encoded)
churn_proportion = churn_counts[True] / total_customers
no_churn_proportion = no_churn_counts[True] / total_customers

print(f"Proporci√≥n de clientes que cancelaron (Churn): {churn_proportion:.2%}")
print(f"Proporci√≥n de clientes que no cancelaron (No Churn): {no_churn_proportion:.2%}")

if abs(churn_proportion - no_churn_proportion) > 0.1: # Threshold for considering imbalance
    print("\nExiste un desbalance significativo entre las clases de clientes que cancelaron y los que no.")
else:
    print("\nLas clases est√°n relativamente balanceadas.")
```

## **Balanceo de Clases** üìù

 La t√©cnica SMOTE ayuda a  balancear las clases. Antes, ten√≠amos una cantidad significativamente menor de clientes con 'Deserci√≥n_S√≠'. Ahora, ambas clases ('Deserci√≥n_S√≠' y 'Deserci√≥n_No') tienen un n√∫mero similar de muestras, lo que es ideal para entrenar modelos predictivos y evitar que se sesguen hacia la clase mayoritaria.

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

# Definir las caracter√≠sticas (X) y la variable objetivo (y)
X = df_encoded.drop(['Deserci√≥n_S√≠', 'Deserci√≥n_No'], axis=1)
y = df_encoded['Deserci√≥n_S√≠'] # La clase minoritaria que queremos balancear

print(f"Distribuci√≥n de clases antes de SMOTE: {Counter(y)}")

sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)

print(f"Distribuci√≥n de clases despu√©s de SMOTE: {Counter(y_res)}")
print(f"Forma del DataFrame X despu√©s de SMOTE: {X_res.shape}")
print(f"Forma del DataFrame y despu√©s de SMOTE: {y_res.shape}")
```

Al revisar las estad√≠sticas descriptivas de X_res, podemos ver las escalas de las caracter√≠sticas num√©ricas como Jubilado, Antiguedad, Cargos_mensuales y Cargos_totales. Si estas caracter√≠sticas tienen rangos de valores muy diferentes, la normalizaci√≥n o estandarizaci√≥n ser√° beneficiosa para algoritmos que son sensibles a la escala de las caracter√≠sticas (por ejemplo, aquellos basados en distancias como SVM, KNN, o redes neuronales).

```python
print('Estad√≠sticas descriptivas de las caracter√≠sticas num√©ricas en X_res:')
display(X_res.describe())
```

A continuaci√≥n, aplicaremos la estandarizaci√≥n a las caracter√≠sticas num√©ricas para que todas est√©n en una escala similar, lo que es beneficioso para la mayor√≠a de los modelos de machine learning.

```python
from sklearn.preprocessing import StandardScaler

# Identificar las columnas num√©ricas para estandarizar
# Excluimos las columnas booleanas que fueron creadas por one-hot encoding y 'Jubilado' que ya es binaria

# Obtener los nombres de las columnas que no son booleanas ni 'Jubilado'
# Las columnas booleanas son las que terminan en '_S√≠', '_No', etc.

# Seleccionar solo las columnas num√©ricas de X_res que no son binarias (Jubilado ya es 0 o 1)
numeric_cols = ['Antiguedad', 'Cargos_mensuales', 'Cargos_totales']

# Crear una copia de X_res para aplicar la estandarizaci√≥n
X_res_scaled = X_res.copy()

# Inicializar el estandarizador
scaler = StandardScaler()

# Aplicar estandarizaci√≥n solo a las columnas num√©ricas seleccionadas
X_res_scaled[numeric_cols] = scaler.fit_transform(X_res_scaled[numeric_cols])

print('Estad√≠sticas descriptivas de las caracter√≠sticas num√©ricas en X_res despu√©s de la estandarizaci√≥n:')
display(X_res_scaled[numeric_cols].describe())
print('\nPrimeras 5 filas del DataFrame X_res_scaled:')
display(X_res_scaled.head())
```

 # üéØ **Correlaci√≥n y Selecci√≥n de Variables**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Combinar X_res_scaled y y_res para calcular la correlaci√≥n con la variable objetivo
df_balanced = X_res_scaled.copy()
df_balanced['Deserci√≥n_S√≠'] = y_res

# Calcular la matriz de correlaci√≥n
correlation_matrix = df_balanced.corr()

# Visualizar la matriz de correlaci√≥n
plt.figure(figsize=(20, 15))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlaci√≥n de todas las Variables (despu√©s de SMOTE y estandarizaci√≥n)')
plt.show()
```

Aqu√≠ tenemos la matriz de correlaci√≥n visualizada en un mapa de calor. . Los valores cercanos a 1 o -1 indican una fuerte correlaci√≥n positiva o negativa, respectivamente, mientras que los valores cercanos a 0 indican poca o ninguna correlaci√≥n. Ahora, para enfocarnos en la deserci√≥n, veamos las correlaciones espec√≠ficamente con la variable Deserci√≥n_S√≠.

```python
# Mostrar la correlaci√≥n de todas las caracter√≠sticas con la variable objetivo 'Deserci√≥n_S√≠'
churn_correlations = correlation_matrix['Deserci√≥n_S√≠'].sort_values(ascending=False)

print('\nCorrelaci√≥n de las variables con la Deserci√≥n (Deserci√≥n_S√≠):')
display(churn_correlations)
```

Esta lista se muestra qu√© variables tienen la correlaci√≥n m√°s fuerte (positiva o negativa) con la deserci√≥n. Las variables con los valores absolutos m√°s altos son las que probablemente tienen un mayor impacto en la predicci√≥n del churn y ser√°n cruciales para el modelo.

Relaci√≥n entre Antig√ºedad (Tiempo de Contrato) y Deserci√≥n
Aqu√≠, visualizaremos c√≥mo la Antiguedad (que representa el tiempo de contrato o la antig√ºedad del cliente) se relaciona con la Deserci√≥n. Un boxplot nos mostrar√° la distribuci√≥n de la antig√ºedad para los clientes que cancelaron (Deserci√≥n_S√≠ = True) y para los que no (Deserci√≥n_S√≠ = False).

```python

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(x='Deserci√≥n_S√≠', y='Antiguedad', data=df_balanced, palette='viridis', hue='Deserci√≥n_S√≠', legend=False)
plt.title('Antig√ºedad del Cliente vs. Deserci√≥n')
plt.xlabel('Deserci√≥n (True = S√≠, False = No)')
plt.ylabel('Antig√ºedad (Estandarizada)')
plt.show()

```

### Relaci√≥n entre Cargos Totales y Deserci√≥n

Ahora, analizaremos c√≥mo los `Cargos_totales` (el gasto total del cliente) se relacionan con la `Deserci√≥n`. Un boxplot similar nos ayudar√° a entender si hay diferencias significativas en el gasto total entre los clientes que cancelan y los que permanecen activos.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Deserci√≥n_S√≠', y='Cargos_totales', data=df_balanced, palette='plasma', hue='Deserci√≥n_S√≠', legend=False)
plt.title('Cargos Totales del Cliente vs. Deserci√≥n')
plt.xlabel('Deserci√≥n (True = S√≠, False = No)')
plt.ylabel('Cargos Totales (Estandarizados)')
plt.show()
```

 # **Modelado Predictivo** ü§ñ


Iniciamos con la Separaci√≥n de los datos para ir entrenando el modelo de predicci√≥n

```python
from sklearn.model_selection import train_test_split

# Dividir el conjunto de datos en entrenamiento y prueba
# X_res_scaled son las caracter√≠sticas estandarizadas y y_res es la variable objetivo balanceada
X_train, X_test, y_train, y_test = train_test_split(X_res_scaled, y_res, test_size=0.3, random_state=42, stratify=y_res)

print(f"Dimensiones de X_train: {X_train.shape}")
print(f"Dimensiones de X_test: {X_test.shape}")
print(f"Dimensiones de y_train: {y_train.shape}")
print(f"Dimensiones de y_test: {y_test.shape}")

print(f"\nDistribuci√≥n de clases en y_train: {y_train.value_counts(normalize=True)}")
print(f"Distribuci√≥n de clases en y_test: {y_test.value_counts(normalize=True)}")
```

Entrenamos un modelo de Regresi√≥n Log√≠stica y un modelo de √Årbol de Decisi√≥n utilizando X_train y y_train. Luego, evaluamos ambos modelos con X_test y y_test, comparando su rendimiento y discutiendo el impacto de la normalizaci√≥n.

## Entrenar Modelo de Regresi√≥n Log√≠stica

Entrenar un modelo de Regresi√≥n Log√≠stica utilizando los datos de entrenamiento escalados (`X_train`, `y_train`). La Regresi√≥n Log√≠stica se beneficia de los datos normalizados para asegurar un c√°lculo de coeficientes adecuado.


```python
from sklearn.linear_model import LogisticRegression

# Crear una instancia del modelo de Regresi√≥n Log√≠stica
log_reg_model = LogisticRegression(random_state=42, solver='liblinear') # Usamos liblinear por ser bueno con datasets peque√±os y binarios

# Entrenar el modelo con los datos de entrenamiento escalados
log_reg_model.fit(X_train, y_train)

print("Modelo de Regresi√≥n Log√≠stica entrenado exitosamente.")
```

## Evaluar Modelo de Regresi√≥n Log√≠stica

Evaluamos el rendimiento del modelo de Regresi√≥n Log√≠stica en los datos de prueba (X_test, y_test) utilizando m√©tricas como precisi√≥n, recall, puntuaci√≥n F1 y exactitud. Esto demostrar√° el impacto de la normalizaci√≥n en un modelo basado en la distancia.


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Hacer predicciones en el conjunto de prueba
y_pred_log_reg = log_reg_model.predict(X_test)

# Calcular las m√©tricas de evaluaci√≥n
accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
precision_log_reg = precision_score(y_test, y_pred_log_reg)
recall_log_reg = recall_score(y_test, y_pred_log_reg)
f1_log_reg = f1_score(y_test, y_pred_log_reg)

# Imprimir los resultados
print("M√©tricas de Evaluaci√≥n para el Modelo de Regresi√≥n Log√≠stica:")
print(f"  Accuracy: {accuracy_log_reg:.4f}")
print(f"  Precision: {precision_log_reg:.4f}")
print(f"  Recall: {recall_log_reg:.4f}")
print(f"  F1-Score: {f1_log_reg:.4f}")
```

## Entrenar Modelo de √Årbol de Decisi√≥n


Entrenar un modelo de √Årbol de Decisi√≥n utilizando los datos de entrenamiento (`X_train`, `y_train`). A diferencia de la Regresi√≥n Log√≠stica, los √°rboles de decisi√≥n son menos sensibles a la escala de las caracter√≠sticas, pero la estandarizaci√≥n previa no les perjudica y puede ser √∫til si se consideran otros modelos.



```python
from sklearn.tree import DecisionTreeClassifier

# Crear una instancia del modelo de √Årbol de Decisi√≥n
dt_model = DecisionTreeClassifier(random_state=42)

# Entrenar el modelo con los datos de entrenamiento
dt_model.fit(X_train, y_train)

print("Modelo de √Årbol de Decisi√≥n entrenado exitosamente.")
```

## Evaluar Modelo de √Årbol de Decisi√≥n


Evaluar el rendimiento del modelo de √Årbol de Decisi√≥n en los datos de prueba (`X_test`, `y_test`) utilizando m√©tricas como precisi√≥n, recall, puntuaci√≥n F1 y exactitud.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Hacer predicciones en el conjunto de prueba
y_pred_dt = dt_model.predict(X_test)

# Calcular las m√©tricas de evaluaci√≥n
accuracy_dt = accuracy_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt)
recall_dt = recall_score(y_test, y_pred_dt)
f1_dt = f1_score(y_test, y_pred_dt)

# Imprimir los resultados
print("M√©tricas de Evaluaci√≥n para el Modelo de √Årbol de Decisi√≥n:")
print(f"  Accuracy: {accuracy_dt:.4f}")
print(f"  Precision: {precision_dt:.4f}")
print(f"  Recall: {recall_dt:.4f}")
print(f"  F1-Score: {f1_dt:.4f}")
```

La estandarizaci√≥n, que aplicamos a las columnas num√©ricas (Antiguedad, Cargos_mensuales, Cargos_totales), asegura que los modelos basados en distancia o en optimizaci√≥n de par√°metros no se sesguen por la magnitud de las variables de la siguiente manera:

Modelos Basados en Distancia (como K-Nearest Neighbors, SVM con kernel lineal):

Estos modelos calculan la distancia entre los puntos de datos para determinar similitudes o clasificaciones. Si una variable tiene un rango de valores mucho mayor que otra (por ejemplo, Cargos_totales puede ir de 0 a miles, mientras que Antiguedad va de 0 a 72), su magnitud puede dominar el c√°lculo de la distancia.
Al estandarizar, todas las caracter√≠sticas num√©ricas se transforman para tener una media de 0 y una desviaci√≥n est√°ndar de 1. Esto significa que todas las caracter√≠sticas contribuyen por igual a la distancia, evitando que una variable con valores m√°s grandes tenga un peso desproporcionado.

## **Usando Matriz de Confusion**

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Matriz de Confusi√≥n para Regresi√≥n Log√≠stica
cm_log_reg = confusion_matrix(y_test, y_pred_log_reg)

plt.figure(figsize=(6, 5))
sns.heatmap(cm_log_reg, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['No Deserci√≥n', 'Deserci√≥n'], yticklabels=['No Deserci√≥n', 'Deserci√≥n'])
plt.xlabel('Predicho')
plt.ylabel('Verdadero')
plt.title('Matriz de Confusi√≥n - Regresi√≥n Log√≠stica')
plt.show()
```

```python
# Matriz de Confusi√≥n para √Årbol de Decisi√≥n
cm_dt = confusion_matrix(y_test, y_pred_dt)

plt.figure(figsize=(6, 5))
sns.heatmap(cm_dt, annot=True, fmt='d', cmap='Greens', cbar=False,
            xticklabels=['No Deserci√≥n', 'Deserci√≥n'], yticklabels=['No Deserci√≥n', 'Deserci√≥n'])
plt.xlabel('Predicho')
plt.ylabel('Verdadero')
plt.title('Matriz de Confusi√≥n - √Årbol de Decisi√≥n')
plt.show()
```

```python
#Ajustamos el Modelo Arbol de Decision para controlar el overfitting.

from sklearn.tree import DecisionTreeClassifier

# Crear una instancia del modelo de √Årbol de Decisi√≥n con max_depth para controlar el overfitting
dt_model = DecisionTreeClassifier(random_state=42, max_depth=5) # Ajustamos max_depth a 5

# Entrenar el modelo con los datos de entrenamiento
dt_model.fit(X_train, y_train)

print("Modelo de √Årbol de Decisi√≥n reentrenado exitosamente con max_depth=5.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Hacer predicciones en el conjunto de prueba con el modelo ajustado
y_pred_dt = dt_model.predict(X_test)

# Calcular las m√©tricas de evaluaci√≥n
accuracy_dt = accuracy_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt)
recall_dt = recall_score(y_test, y_pred_dt)
f1_dt = f1_score(y_test, y_pred_dt)

# Imprimir los resultados
print("M√©tricas de Evaluaci√≥n para el Modelo de √Årbol de Decisi√≥n (con max_depth=5):")
print(f"  Accuracy: {accuracy_dt:.4f}")
print(f"  Precision: {precision_dt:.4f}")
print(f"  Recall: {recall_dt:.4f}")
print(f"  F1-Score: {f1_dt:.4f}")
```

###An√°lisis del Impacto de max_depth=5:

Accuracy (Exactitud): Disminuy√≥ ligeramente de 0.8131 a 0.8031. Esto indica que el modelo ahora clasifica un porcentaje ligeramente menor de las muestras correctamente en general.

Precision (Precisi√≥n): Aument√≥ de 0.7965 a 0.8145. Una mayor precisi√≥n significa que, cuando el modelo predice 'Deserci√≥n', es m√°s probable que esa predicci√≥n sea correcta. Esto sugiere que el modelo es ahora m√°s cauteloso o tiene menos falsos positivos en sus predicciones de deserci√≥n.

Recall (Exhaustividad): Disminuy√≥ notablemente de 0.8412 a 0.7850. Una disminuci√≥n en el recall indica que el modelo ahora identifica un porcentaje menor de los clientes que realmente van a desertar. Es decir, hay m√°s falsos negativos.

F1-Score: Disminuy√≥ de 0.8182 a 0.7995. El F1-Score es la media arm√≥nica de precisi√≥n y recall, por lo que su disminuci√≥n refleja el trade-off entre estas dos m√©tricas.
